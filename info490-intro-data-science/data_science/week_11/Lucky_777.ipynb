```python
import requests
def read_remote(url):
  # assumes the url is already encoded (see urllib.parse.urlencode)
  with requests.get(url) as response:
    response.encoding = 'utf-8'
    if response.status_code == requests.codes.ok: # (that is 200)
      return response.text
  return None

HAMLET_URL = "https://www.gutenberg.org/cache/epub/2265/pg2265.txt"
hamlet = read_remote(HAMLET_URL)
```

```python
def find_lucky(text, num):
  import re

  pattern = r"[a-z]+'?[a-z]?"
  regex = re.compile(pattern, re.IGNORECASE)
  tokens = regex.findall(text)
  normalized_tokens = [token.lower() for token in tokens]

  # checks the number of characters (criteria 1)
  potential_words = []
  for token in normalized_tokens:
    if len(token) == num and token not in potential_words:
      potential_words.append(token)

  # check that each word shows up a 'num' number of times in the text (criteria 2)
  lucky = []
  num_check = 0
  for word in potential_words:
    for token in normalized_tokens:
      if word == token:
        num_check += 1

    if num_check == num: # check there are only 'num' appearances of the word
      lucky.append(word)
      num_check = 0
    else:
      num_check = 0
  
  if len(lucky) == num: # check that each word had the correct number of appearances (criteria 3)
    return sorted(lucky)
  else:
    return []

text = "I like to eat like and do's to or or and Don't"

find_lucky(text,2)
```
